```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from konlpy.tag import Okt
import re
```

```python
def estimate(review):
    ## 모델 로드해오기
    review = "싫어요"

    loaded_model = keras.models.load_model('model_best.h5')

    stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다','너무']

    okt = Okt()
    tokenizer = Tokenizer()

    ### 리뷰데이터 전처리
    max_len = 30
    ### 한글과 공백 제외 제거
    review = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','', review)
    review = okt.morphs(review, stem=True) # 토큰화
    review = [word for word in review if not word in stopwords] # 불용어 제거
    encoded = tokenizer.texts_to_sequences([review]) # 정수 인코딩
    pad_new = pad_sequences(encoded, maxlen = max_len) # 패딩
    score = float(loaded_model.predict(pad_new)) # 예측

    if score > 0.5:
        return 1
        #print(1)
    else :
        return 0
    #print(0)
    ## review를 모델에 넣어서 값 얻기
    ## 0또는 1 반환하기
```

```python
### [질문]이 데이터프레임의 request에 포함되어 있는지 확인하는 기능을 구현
# - 있으면 : 데이터프레임에서 해당 index의 response값으로 응답처리
# - 없으면 : "무슨 말인지 잘 모르겠네요~" 응답처리

# ** 챗봇은 질문자와 반복해서 주고 받아야하기때문에 질문 처리기능이 함수로 만들어져있어야함


###  맨처음 챗봇 -> 정보를 넣으면 알려주는 것이므로  request 로 정보 종류에 대해 받아야함
def chat_prof(req, professor_name, review=""):
    df = pd.read_csv("교수정보.csv")
    
    for i in range(len(df)):
        if df["교수명"][i] == professor_name:
            val = i
            break
            
    if req == "전공":
        return "{} 교수님의 전공은 {}입니다.".format(professor_name, df["전공"][val])
    elif req == "연구실":
        return "{} 교수님의 연구실은 {}입니다.".format(professor_name, df["연구실"][val])
    elif req == "이메일":
        return "{} 교수님의 이메일은 {}입니다.".format(professor_name, df["이메일"][val])
    elif req == "전화번호":
        return "{} 교수님의 전화번호는 {}입니다.".format(professor_name, df["전화번호"][val])
    elif req == "강의평":
        return "{} 교수님의 강의평은 좋아요 {}, 별로에요 {} 입니다.".format(professor_name, df["좋아요"][val], df["별로에요"][val])
    elif req == "강의평가하기":
        num = estimate(review)
        # globals()['reviews_{}'.format(val)].append(review) 리스트로 저장

        df["좋아요" if num == 1 else "별로에요"][val] += 1
        ### 여기에 감성평가함수가 들어가서 새로운 강의평가가 들어올 때 마다 새로운 리스트를 받아서 강의평(호감도)를 반환하여 df["강의평"] 컬럼에 저장
        ### df["강의평"][val] = 강의평가(reviews)
        ###  교수별 강의평가 리스트를 데이터프레임으로 만들기
        ### df_review = pd.DataFrame(globals()['reviews_{}'.format(val)])
        ### 각각의 강의평가 문자열로 이루어진 교수별 리스트 csv파일로 저장
        ### df_review.to_csv("reviews_{}".format(val))
        df.to_csv("교수정보.csv")
        
        return "저장되었습니다."
    
    elif req in chat_dic.values():
        return req
    else :
        return "'전공'/'연구실'/'이메일'/'전화번호'/'강의평'/'강의평가하기' 만 입력해주세요."
```

```python
### 챗봇 인터페이스 만들기
# - 질문자와 응답챗봇간의 인터페이스
# - 질무자의 질문을 함수에 전달
# - 함수가 리턴한 응답을 질문자에게 보여주기
# - 인터페이스 종료까지

professor_name = ""
#무한반복시키기
while True :
    ### 외부장치 오류 있을 수 있으므로 try, except로 
    try :
        if not professor_name: # 전역 변수가 비어있으면 교수님 이름 입력받기
            req = input("궁금한 교수님 성함을 입력해주세요: ")
            if req in chat_dic.values(): # 입력받은 교수님 이름이 chat_dic.values() 경우에만 전역 변수에 저장합니다.
                professor_name = req
            elif req == "exit":
                print("감사합니다")
                break
            else : 
                req = input("다시 입력해주세요: ")
                if req == "exit":
                    print("감사합니다")
                    break
                
        else: # 전역 변수가 비어있지 않은 경우에는 해당 교수님 정보를 출력합니다.    
            req = input("{} 교수님에 대해 더 알고싶은 정보를 입력해 주세요 (전공/연구실/이메일/전화번호/강의평/강의평가하기) \
                                    (다른 교수님을 원하시면 성함을, 더 알고싶은 정보가 없으면 '없어'를 입력해주세요): ".format(professor_name))
            
            if req == "없어":
                print("감사합니다")
                break
            ### 질문을 함수에 보내서 응답메세지 받아오기
            elif req in chat_dic.values():
                professor_name = req
            elif req == "강의평가하기":
                review = input("강의평을 자유롭게 입력해주세요 : ")
                ### 원본 Dataframe의 일부를 복사하거나 인덱싱 후 값을 수정할 때 발생할 수 있는 "SettingWithCopyWarning or "SettingWithCopyError" 암시적으로 끄기
                pd.set_option('mode.chained_assignment',  None)
                print("ChatBot : ",chat_prof(req, professor_name, review))
            else :
                print("ChatBot : ", chat_prof(req, professor_name))
                
    except :
        print("알 수 없는 오류가 발생했습니다. 종료합니다.")
        break
```
