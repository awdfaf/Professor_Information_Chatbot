```python
# 패키지
from selenium import webdriver
from selenium.webdriver.common.by import By

# 크롬드라이버 다운 - 버전 주의
# 프로그램에서 실행

driver = webdriver.Chrome('chrome_109_driver.exe')

# url 주소 전달
# 정융대 입장
url = "https://itc.pknu.ac.kr/html/00_main/"
driver.get(url)
driver.maximize_window()

# 교수진 페이지
Button_1 = driver.find_element(By.XPATH,
    '//*[@id="gnb1m3"]')
Button_1.click()

a = driver.find_element(By.XPATH,
    '//*[@id="container"]/div[2]/ul')
b = a.find_elements(By.TAG_NAME,"li")

majors = []
names = []
phones = []
labs = []
emails = []

for i in range(1,len(b)+1):
    # 교수 정보 얻기
    Button = driver.find_element(By.XPATH,
        '//*[@id="container"]/div[2]/ul/li[{}]/a'.format(i))
    Button.click()
    
    dd = driver.find_element(By.XPATH,
                         '//*[@id="container"]/div[2]/h3')
    
    big_table = driver.find_elements(By.XPATH,
        '//*[@id="container"]/div[2]/div')

    for human in big_table:
        #print(human.text)
        a = []
        info = human.text.split("\n")
    major = []
    major.append(dd.text)
    name = []
    phone = []
    lab = []
    email = []

    for temp in info :

        if "교수" in temp.split(" ") :
            name.append(temp)
        if "051" in temp.split("-") :
            phone.append(temp)
        if "연구실" in temp.split(" ") and "홈페이지" not in temp.split(" ") and "·" not in temp.split(" "):
            lab.append(temp)
        if "pknu.ac.kr" in temp.split("@") :
            email.append(temp)
    
    majors.append(major)
    names.append(name)
    phones.append(phone)
    labs.append(lab)
    emails.append(email)
    
    
    try:
        professor = {}
        for i in range(len(name)) :
            professor[name[i]] = [phone[i],lab[i],email[i]]

        print(professor)
        
    except:
        continue



    
    
# 
#print(name)
#print()

# # 통계데이터사이언스
# Button_2 = driver.find_element(By.XPATH,
#     '//*[@id="container"]/div[2]/ul/li[2]/a')
# Button_2.click()



# a = tablebody.find_element(By.XPATH,
#     '//*[@id="container"]/div[3]/article[{}]'.format(i))
#     b = a.find_element(By.TAG_NAME,
#     'p')
#     titles.append(b.text)
# b = a.find_element(By.TAG_NAME,
#     'p')
#     titles.append(b.text)


# id = driver.find_element(By.NAME,'memId')
# password = driver.find_element(By.NAME,'memPwd')

# id.send_keys('201811723')
# password.send_keys('awdfaf0409')

# loginButton_2 = driver.find_element(By.XPATH,
#     '//*[@id="frmLogin"]/button')
# loginButton_2.click()

# loginButton_3 = driver.find_element(By.XPATH,
#     '//*[@id="main"]/div[4]/div/section/div[2]/ul[2]/li[2]/a')
# loginButton_3.click()

# driver.get("https://www.pknu.ac.kr/main/378")

# newboard = driver.find_element(By.XPATH,
#     '//*[@id="submenu"]/div/div[2]/ul/li[4]/a')
# newboard.click()

# tablebody =driver.find_element(By.XPATH,
#     '//*[@id="container"]')

# titles = []


# for i in range(1,21):
#     a = tablebody.find_element(By.XPATH,
#     '//*[@id="container"]/div[3]/article[{}]'.format(i))
#     b = a.find_element(By.TAG_NAME,
#     'p')
#     titles.append(b.text)

# nextbt = driver.find_element(By.XPATH,
#     '//*[@id="container"]/div[3]/div[2]/a')
# nextbt.click()
```
